## –û—Å–Ω–æ–≤–Ω—ã–µ —Ç–µ–æ—Ä–µ—Ç–∏—á–µ—Å–∫–∏–µ –ø–æ–ª–æ–∂–µ–Ω–∏—è
1. –õ–∏–Ω–µ–π–Ω–∞—è —Ä–µ–≥—Ä–µ—Å—Å–∏–æ–Ω–Ω–∞—è –º–æ–¥–µ–ª—å

–õ–∏–Ω–µ–π–Ω–∞—è —Ä–µ–≥—Ä–µ—Å—Å–∏—è –Ω–µ–∫–æ—Ç–æ—Ä–æ–π –∑–∞–≤–∏—Å–∏–º–æ–π –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π y –Ω–∞ –Ω–∞–±–æ—Ä –Ω–µ–∑–∞–≤–∏—Å–∏–º—ã—Ö –ø–µ—Ä–µ–º–µ–Ω–Ω—ã—Ö  x = (x_1, ‚Ä¶, x_n), –≥–¥–µ n ‚Äì —ç—Ç–æ —á–∏—Å–ª–æ –Ω–∞–±–ª—é–¥–µ–Ω–∏–π, –ø—Ä–µ–¥–ø–æ–ª–∞–≥–∞–µ—Ç, —á—Ç–æ –ª–∏–Ω–µ–π–Ω–æ–µ –æ—Ç–Ω–æ—à–µ–Ω–∏–µ –º–µ–∂–¥—É y –∏ x: $y =\beta_0  + \beta_1x_1 + ‚ãØ + \beta_nx_n + ùúÄ$ - —É—Ä–∞–≤–Ω–µ–Ω–∏–µ —Ä–µ–≥—Ä–µ—Å—Å–∏–∏. $\beta_0, \beta_1, ‚Ä¶, \beta_n$ ‚Äì –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç—ã —Ä–µ–≥—Ä–µ—Å—Å–∏–∏ (–ø–∞—Ä–∞–º–µ—Ç—Ä—ã), –∏ ùúÄ ‚Äì —Å–ª—É—á–∞–π–Ω–∞—è –æ—à–∏–±–∫–∞.

2. –ú–ù–ö (–ú–µ—Ç–æ–¥ –ù–∞–∏–º–µ–Ω—å—à–∏—Ö –ö–≤–∞–¥—Ä–∞—Ç–æ–≤)

$RSS(\beta)=\left\| Y-X^{\intercal}\beta \right\|=(Y-X^{\intercal}\beta)^{\intercal}\cdot(Y-X^{\intercal}\beta)\to min\beta$

–ù–æ—Ä–º–∞–ª—å–Ω—ã–µ —É—Ä–∞–≤–Ω–µ–Ω–∏—è: $XX^{\intercal}\beta=XY$

–†–µ—à–µ–Ω–∏–µ —Å–∏—Å—Ç–µ–º—ã –Ω–æ—Ä–º–∞–ª—å–Ω—ã—Ö —É—Ä–∞–≤–Ω–µ–Ω–∏–π –≤ —Ä–µ–≥—É–ª—è—Ä–Ω–æ–º —Å–ª—É—á–∞–µ: $\widehat{\beta}=(XX^{\intercal})^{-1}XY$

$RSS(\beta)=\sum_{i=1}^{n}\left( Y_{i}-\sum_{j=1}^{m}x_{ji}\beta_{j} \right)^{2}$

$\frac{\partial RSS}{\partial \beta_{k}} = -2\sum_{i=1}^{n}x_{ki}\left(Y_{i}- \sum_{j=1}^{m}x_{ji}\beta_{j}\right)=0, k=\overline{1,m}$

3. –ù–µ—Å–º–µ—â—ë–Ω–Ω–∞—è –æ—Ü–µ–Ω–∫–∞ –¥–∏—Å–ø–µ—Ä—Å–∏–∏ $(S^{2})$

$S^{2} = \frac{SS_{e}}{n-r} = \frac{RSS(\widehat{\beta})}{n-r}$

$SS_{e}=RSS(\widehat{\beta})$, –∞ r-—Ä–∞–Ω–≥ –º–∞—Ç—Ä–∏—Ü—ã X

4. –î–æ–≤–µ—Ä–∏—Ç–µ–ª—å–Ω—ã–π –∏–Ω—Ç–µ—Ä–≤–∞–ª

–î–ù–û —Ñ—É–Ω–∫—Ü–∏–∏:$\psi=—Å^{\intercal}\beta$

–ú–ù–ö-–æ—Ü–µ–Ω–∫–∞: $\widehat{\psi}=—Å^{\intercal}\beta$

–≠–ª–µ–º–µ–Ω—Ç –¥–∏—Å–ø–µ—Ä—Å–∏–∏ –æ—Ü–µ–Ω–∫–∏: $b_{\psi}=c^{\intercal}(XX^{\intercal})^{-1}c$

–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ $\widehat{\psi}\sim \mathcal{N}(\psi,\sigma^{2}b)$

$\frac{\widehat{\psi}-\psi}{S\sqrt{b}}\sim \mathcal{S}_{n-r}$

–î.–ò. —É—Ä–æ–≤–Ω—è –¥–æ–≤–µ—Ä–∏—è $1-\alpha: \left[ \widehat{\psi} - X_{\alpha}S\sqrt{b};\widehat{\psi} + X_{\alpha}S\sqrt{b} \right], –≥–¥–µ X_{\alpha}: \mathcal{S}_{n-r}(X_{\alpha})=1-\frac{\alpha}{2}$

5. –£—Ä–∞–≤–Ω–µ–Ω–∏–µ $A_{\alpha}=\left\{ \overrightarrow{X}:(\overrightarrow{X}-\widehat{\psi})^{\intercal}B^{-1}(\overrightarrow{X}-\widehat{\psi})\le qS^{2}X_{\alpha} \right\}$ –ø—Ä–µ–¥—Å—Ç–∞–≤–ª—è–µ—Ç —Å–æ–±–æ–π —É—Ä–∞–≤–Ω–µ–Ω–∏–µ –¥–æ–≤–µ—Ä–∏—Ç–µ–ª—å–Ω–æ–≥–æ —ç–ª–ª–∏–ø—Å–æ–∏–¥–∞. –ó–¥–µ—Å—å:

- $\widehat{\psi}$ - –≤–µ–∫—Ç–æ—Ä –æ—Ü–µ–Ω–æ–∫ –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤,
- $B$ - –º–∞—Ç—Ä–∏—Ü–∞ –æ—Ü–µ–Ω–æ–∫ –∫–æ–≤–∞—Ä–∏–∞—Ü–∏–∏,
- $S^{2}$ - –æ—Ü–µ–Ω–∫–∞ –¥–∏—Å–ø–µ—Ä—Å–∏–∏,
- q - –ø–æ—Ä–æ–≥–æ–≤–∞—è —Ç–æ—á–∫–∞, —Å–≤—è–∑–∞–Ω–Ω–∞—è —Å —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ–º —Ö–∏-–∫–≤–∞–¥—Ä–∞—Ç —Å \(n-r\) —Å—Ç–µ–ø–µ–Ω—è–º–∏ —Å–≤–æ–±–æ–¥—ã,
- $X_{\alpha}$ - –∑–Ω–∞—á–µ–Ω–∏–µ —Ñ—É–Ω–∫—Ü–∏–∏ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è —Ö–∏-–∫–≤–∞–¥—Ä–∞—Ç —Å (n-r) —Å—Ç–µ–ø–µ–Ω—è–º–∏ —Å–≤–æ–±–æ–¥—ã –∏ —É—Ä–æ–≤–Ω–µ–º –¥–æ–≤–µ—Ä–∏—è $1-\alpha$.

6. –ö—Ä–∏—Ç–µ—Ä–∏–π $\chi^2$ –¥–ª—è –ø—Ä–æ–≤–µ—Ä–∫–∏ –≥–∏–ø–æ—Ç–µ–∑

–æ—Å–Ω–æ–≤–∞–Ω –Ω–∞ —Å–ø—Ä–∞–≤–Ω–µ–Ω–∏–∏ —Ç–µ–æ—Ä–∏—Ç–∏—á–µ—Å–∫–æ–π –ø–ª–æ—Ç–Ω–æ—Å—Ç–∏ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –∏ –≥–∏—Å—Ç–æ–≥—Ä–∞–º–º—ã, –ª–∏–±–æ —Ç–µ–æ—Ä–∏—Ç–∏—á–µ—Å–∫–æ–π –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–Ω–æ–π –º–µ—Ä—ã –∏ –ø–æ–ª–∏–≥–æ–Ω–∞ —á–∞—Å—Ç–æ—Ç.

–†–∞–∑–æ–±—ä–µ–º –º–Ω–æ–∂–µ—Å—Ç–≤–æ –∑–Ω–∞—á–µ–Ω–∏–π —Å–ª. –≤–µ–ª. X –Ω–∞ $N$ –∏–Ω—Ç–µ—Ä–≤–∞–ª–æ–≤ $I_{1}, ..., I_{N}:$ $I_{i} = (a_{i-1}, a_i], i = \overline {1,N}$. –¥–∞–Ω–Ω—ã–µ –∏–Ω—Ç–µ—Ä–≤–∞–ª—ã –Ω–µ –ø–µ—Ä–µ—Å–µ–∫–∞—é—Ç—Å—è –∏ –ø–æ–∫—Ä—ã–≤–∞—é—Ç –≤—Å—ë –º–Ω–æ–∂–µ—Å—Ç–≤–æ –≤–æ–∑–º–æ–∂–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π —Å–ª. –≤–µ–ª. X.

–û–±–æ–∑–Ω–∞—á–∏–º —á–µ—Ä–µ–∑ $p_{i}$ —Ç–µ–æ—Ä–∏—Ç–∏—á–µ—Å–∫—É—é –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å –ø–æ–ø–∞–¥–∞–Ω–∏—è —Å–ª. –≤–µ–ª. X –≤ –∏–Ω—Ç–µ—Ä–≤–∞–ª $I_{i}$, –ø—Ä–∏ —É—Å–ª–æ–≤–∏–∏, —á—Ç–æ —Å–ø—Ä–∞–≤–µ–¥–ª–∏–≤–∞ $H_{0}$.

$p_{i} = P(X \in I_{i}|H_{0})$

–ß–µ—Ä–µ–∑ $n_{i}$ –æ–±–æ–∑–Ω–∞—á–∏–º —á–∏—Å–ª–æ –Ω–∞–±–ª—é–¥–µ–Ω–∏–π –Ω–∞—à–µ–π –≤—ã–±–æ—Ä–∫–∏, –ø–æ–ø–∞–≤—à–∏—Ö –≤ —Å–æ–æ—Ç–≤–µ—Ç–≤–µ—Ç—Å—Ç–≤—É—é—â–∏–π –∏–Ω—Ç–µ—Ä–≤–∞–ª $I_{i}$.

$n_{i} = num(x_{j} \in I_{i})$

–°—Ç–∞—Ç–∏—Å—Ç–∏–∫–∞ –∫—Ä–∏—Ç–µ—Ä–∏—è –æ—Å–Ω–æ–≤–∞–Ω–∞ –Ω–∞ —Å—Ä–∞–≤–Ω–µ–Ω–∏–∏ –æ—Ç–Ω–æ—Å–∏—Ç–µ–ª—å–Ω–æ–π —á–∞—Å—Ç–æ—Ç—ã –∏ —Ç–µ–æ—Ä–∏—Ç–∏—á–µ—Å–∫–æ–π.

$\chi^{2}_{–Ω–∞–±–ª} = \sum \limits_{i=1}^{N} \frac{n}{p_{i}} \bigg ( \frac{n_{i}}{n} - p_{i} \bigg )^{2} = \sum \limits_{i=1}^{N} \frac{(n_{i} - p_{i} n)^{2}}{n p_{i}} \sim \chi^{2}_{N-1}$

–ï—Å–ª–∏ $\mathcal{X}^{2}_{N-1,1-\alpha_{1}} \lt \chi^2_{–Ω–∞–±–ª}$, —Ç–æ –≥–∏–ø–æ—Ç–µ–∑—É –æ—Ç–≤–µ—Ä–≥–∞–µ–º

–ï—Å–ª–∏ $\mathcal{X}^{2}_{N-1,1-\alpha_{1}} \gt \chi^2_{–Ω–∞–±–ª}$, —Ç–æ –≥–∏–ø–æ—Ç–µ–∑—É –ø—Ä–∏–Ω–∏–º–∞–µ–º

7. —Ç–µ–æ—Ä–µ–º–∞ –ö–æ–ª–º–æ–≥–æ—Ä–æ–≤–∞

$D_{n}(\overrightarrow{X}) = \overset{}{\underset{X \in \mathbb R}{sup}} |F_{0}(X_{i}) - F_{n}(X_{i})|$

–ï—Å–ª–∏ –≥–∏–ø–æ—Ç–µ–∑–∞ $H_{0}$ –≤–µ—Ä–Ω–∞, –∏ $F_{0}(X)$ - –Ω–µ–ø—Ä–µ—Ä—ã–≤–Ω–∞—è —Ñ-—è –Ω–∞ $\mathbb R$, —Ç–æ–≥–¥–∞ –∏–º–µ–µ—Ç –º–µ—Å—Ç–æ —Å—Ö–æ–¥–∏–º–æ—Å—Ç—å

$P(\sqrt{n} D_{n}(\overrightarrow{X}) < z) \xrightarrow[n \rightarrow \infty]{} K(z)$, –≥–¥–µ

$K(z)=1-\alpha$ - —Ñ-—è —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –ö–æ–ª–º–æ–≥–æ—Ä–æ–≤–∞.

–ï—Å–ª–∏ $D_{n} \lt z$, —Ç–æ –≥–∏–ø–æ—Ç–µ–∑—É –ø—Ä–∏–Ω–∏–º–∞–µ–º

–ï—Å–ª–∏ $D_{n} \gt z$, —Ç–æ –≥–∏–ø–æ—Ç–µ–∑—É –æ—Ç–≤–µ—Ä–≥–∞–µ–º

8. $\mathbb{F}$-–∫—Ä–∏—Ç–µ—Ä–∏–π

$\mathbb{F}=\frac{MSR}{MSE}$ - —Å—Ç–∞—Ç–∏—Å—Ç–∏–∫–∞ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è

 –≥–¥–µ $MSR=\frac{SSR}{p}=\frac{\sum_{i=1}^{n}\left(\sum_{j=1}^{n}x_{ji} \beta_j - \overline{Y}\right)^2}{p}$ - —Å—Ä–µ–¥–Ω–µ–∫–≤–∞–¥—Ä–∞—Ç–∏—á–Ω–æ–µ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ –æ–±—ä—è—Å–Ω–µ–Ω–Ω—ã—Ö –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–π

 $MSE = \frac{SS_e}{n-p}=\frac{\sum_{i=1}^{n}\left( Y_{i}-\sum_{j=1}^{m}x_{ji}\beta_{j} \right)^{2}}{n-p}, –≥–¥–µ SS_{e}=RSS(\beta)$ - —Å—Ä–µ–¥–Ω–µ–∫–≤–∞–¥—Ä–∞—Ç–∏—á–Ω–æ–µ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ –æ—Å—Ç–∞—Ç–∫–æ–≤

 –ö—Ä–∏—Ç–∏—á–µ—Å–∫–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ $\mathbb{F}_{–∫—Ä}$-–∫—Ä–∏—Ç–µ—Ä–∏—è —É—Ä–æ–≤–Ω—è –∑–Ω–∞—á–∏–º–æ—Å—Ç–∏ $\alpha$ —Å–º–æ—Ç—Ä–∏–º –ø–æ —Ç–∞–±–ª–∏—Ü–µ –∫—Ä–∏—Ç–∏—á–µ—Å–∫–∏—Ö –∑–Ω–∞—á–µ–Ω–∏–µ –∏–ª–∏ —Å –ø–æ–º–æ—â—å—é –∫–æ–¥–∞ scipy.stats.f.ppf($1-\alpha$, dfn, dfd), –≥–¥–µ dfn-—á–∏—Å–ª–∏—Ç–µ–ª—å —Å—Ç–µ–ø–µ–Ω–µ–π —Å–≤–æ–±–æ–¥—ã(p) dfd-–∑–Ω–∞–º–µ–Ω–∞—Ç–∏–ª—å —Å—Ç–µ–ø–µ–Ω–µ–π —Å–≤–æ–±–æ–¥—ã(n-p)

 $\mathbb{F}_{–∫—Ä} \gt \mathbb{F}$ - –≥–∏–ø–æ—Ç–µ–∑—É –ø—Ä–∏–Ω–∏–º–∞–µ–º

 $\mathbb{F}_{–∫—Ä} \lt \mathbb{F}$ - –≥–∏–ø–æ—Ç–µ–∑—É –æ—Ç–≤–µ—Ä–≥–∞–µ–º

## –ü–æ—Å—Ç–∞–Ω–æ–≤–∫–∞ –∑–∞–¥–∞—á–∏
–†–µ–∑—É–ª—å—Ç–∞—Ç—ã —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–æ–≥–æ —ç–∫—Å–ø–µ—Ä–∏–º–µ–Ω—Ç–∞ –ø—Ä–µ–¥—Å—Ç–∞–≤–ª–µ–Ω—ã –≤ –≤–∏–¥–µ –¥–≤—É—Ö –ø–µ—Ä–µ–º–µ–Ω–Ω—ã—Ö. –¢—Ä–µ–±—É–µ—Ç—Å—è –æ—Ü–µ–Ω–∏—Ç—å —Ö–∞—Ä–∞–∫—Ç–µ—Ä (—Å–ª—É—á–∞–π–Ω–æ–π) –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç–∏ –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π Y
 –æ—Ç –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π X. –ü–æ–ª—É—á–µ–Ω–Ω—ã–µ —Ä–µ–∑—É–ª—å—Ç–∞—Ç—ã —Å–æ–¥–µ—Ä–∂–∞—Ç–µ–ª—å–Ω–æ –ø—Ä–æ–∏–Ω—Ç–µ—Ä–ø—Ä–µ—Ç–∏—Ä–æ–≤–∞—Ç—å.

## –í—ã–ø–æ–ª–Ω–µ–Ω–∏–µ —Ä–∞–±–æ—Ç—ã

### 1.–ü–æ—Å—Ç—Ä–æ–∏—Ç—å –≥—Ä–∞—Ñ–∏—á–µ—Å–∫–∏ —Ä–µ–∑—É–ª—å—Ç–∞—Ç—ã —ç–∫—Å–ø–µ—Ä–∏–º–µ–Ω—Ç–∞. –°—Ñ–æ—Ä–º—É–ª–∏—Ä–æ–≤–∞—Ç—å –ª–∏–Ω–µ–π–Ω—É—é —Ä–µ–≥—Ä–µ—Å—Å–∏–æ–Ω–Ω—É—é –º–æ–¥–µ–ª—å –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π Y –ø–æ –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π X. –ü–æ—Å—Ç—Ä–æ–∏—Ç—å –º–µ—Ç–æ–¥ –Ω–∞–∏–º–µ–Ω—å—à–µ–≥–æ –∫–≤–∞–¥—Ä–∞—Ç–∞ –æ—Ü–µ–Ω–∫–∏ –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤ —Å–¥–≤–∏–≥–∞ $\beta_0$ –∏ –º–∞—Å—à—Ç–∞–±–∞ $\beta_1$.  –ü–æ—Å—Ç—Ä–æ–∏—Ç—å –ø–æ–ª—É—á–µ–Ω–Ω—É—é –ª–∏–Ω–∏—é —Ä–µ–≥—Ä–µ—Å—Å–∏–∏. –û—Ü–µ–Ω–∏—Ç—å –≤–∏–∑—É–∞–ª—å–Ω–æ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–∏–µ –ø–æ–ª—É—á–µ–Ω–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö –∏ –ø–æ—Å—Ç—Ä–æ–µ–Ω–Ω–æ–π –æ—Ü–µ–Ω–∫–∏


–õ–∏–Ω–µ–π–Ω–∞—è —Ä–µ–≥—Ä–µ—Å—Å–∏–æ–Ω–Ω–∞—è –º–æ–¥–µ–ª—å –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π Y –ø–æ –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π X:

$Y = \beta_0 + \beta_1X + \varepsilon,$

–≥–¥–µ $\beta_0$ - —Å–¥–≤–∏–≥, $\beta_1$ - –º–∞—Å—à—Ç–∞–±, –∞ $\varepsilon$ - —Å–ª—É—á–∞–π–Ω–∞—è –æ—à–∏–±–∫–∞.

–î–ª—è –ø–æ–ª—É—á–µ–Ω–∏—è –æ—Ü–µ–Ω–æ–∫ –±—É–¥–µ–º –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å –º–µ—Ç–æ–¥ –Ω–∞–∏–º–µ–Ω—å—à–∏—Ö –∫–≤–∞–¥—Ä–∞—Ç–æ–≤. –í—ã—á–∏—Å–ª–∏–º –æ—Ü–µ–Ω–∫–∏ –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤ $\hat{\beta_0}$ –∏ $\hat{\beta_1}$:

$\hat{\beta_1} = \frac{\sum(X-\overline{X})(Y-\overline{Y})}{\sum(X-\overline{X})^2},$

$\hat{\beta_0} = \overline{Y} - \hat{\beta_1}\overline{X},$

```python
import matplotlib.pyplot as plt
import numpy as np

Y = np.array([16.81, 3.45, 6.39, 1.67, 13.91, 1.11, 6.20, 17.13, 8.57, 9.23, 8.66, 6.01, 9.31, 7.64, 4.30, 5.28, 14.32,1.53, 16.19, 14.03, 21.10, 8.32, 3.65, 3.87, 19.21, 6.20, 19.03, 16.54, 11.21, 5.56, 9.17, 0.79, 9.66, 5.77,4.84, 12.58, 5.91, 1.93, 2.88, 1.02, 2.62, 7.64, 9.86, 6.59, 18.69, 0.81, 15.91, 2.68, 15.63, 14.85])
X = np.array([0, 4, 1, 3, 2, 4, 2, 3, 2, 2, 4, 4, 0, 1, 2, 3, 2, 4, 0, 3, 2, 2, 4, 0, 1, 3, 1, 0, 0, 0, 4, 0, 2, 3,0, 3, 4, 2, 1, 1, 1, 3, 0, 1, 4, 2, 1, 2, 1, 0])
plt.scatter(X, Y)
plt.xlabel("X")
plt.ylabel("Y")
plt.title("–ì—Ä–∞—Ñ–∏–∫ –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç–∏ Y –æ—Ç X")
plt.show()
```

```python
X_mean = np.mean(X)
Y_mean = np.mean(Y)
# –í—ã—á–∏—Å–ª—è–µ–º –æ—Ü–µ–Ω–∫–∏ –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤ —Å–¥–≤–∏–≥–∞ –∏ –º–∞—Å—à—Ç–∞–±–∞ (b_0,b_1)
beta_1 = np.sum((X - X_mean) * (Y - Y_mean)) / np.sum((X - X_mean) ** 2)
beta_0 = Y_mean - beta_1 * X_mean
print("Œ≤1 =",beta_1)
print("Œ≤0 =",beta_0)

# –ü–æ—Å—Ç—Ä–æ–µ–Ω–∏–µ –ª–∏–Ω–∏–∏ —Ä–µ–≥—Ä–µ—Å—Å–∏–∏
plt.scatter(X, Y)
plt.plot([min(X), max(X)], [beta_0 + beta_1 * min(X), beta_0 + beta_1 * max(X)], color='green')
plt.xlabel("X")
plt.ylabel("Y")
plt.title("–õ–∏–Ω–µ–π–Ω–∞—è —Ä–µ–≥—Ä–µ—Å—Å–∏—è")
plt.show()
```

### 2.–ü–æ—Å—Ç—Ä–æ–∏—Ç—å –∏ –∏–Ω—Ç–µ—Ä–ø—Ä–µ—Ç–∏—Ä–æ–≤–∞—Ç—å –Ω–µ—Å–º–µ—â—ë–Ω–Ω—É—é –æ—Ü–µ–Ω–∫—É –¥–∏—Å–ø–µ—Ä—Å–∏–∏. –ù–∞ –±–∞–∑–µ –æ—à–∏–±–æ–∫ –ø–æ—Å—Ç—Ä–æ–∏—Ç—å –≥–∏—Å—Ç–æ–≥—Ä–∞–º–º—É —Å —à–∞–≥–æ–º h. –ü—Ä–æ–≤–µ—Ä–∏—Ç—å –≥–∏–ø–æ—Ç–µ–∑—É –Ω–æ—Ä–º–∞–ª—å–Ω–æ—Å—Ç–∏ –æ—à–∏–±–æ–∫ –Ω–∞ —É—Ä–æ–≤–Ω–µ –∑–Ω–∞—á–∏–º–æ—Å—Ç–∏ –ø–æ $\chi^{2}$. –û—Ü–µ–Ω–∏—Ç—å —Ä–∞—Å—Å—Ç–æ—è–Ω–∏–µ –ø–æ–ª—É—á–µ–Ω–Ω–æ–π –æ—Ü–µ–Ω–∫–∏ –¥–æ –∫–ª–∞—Å—Å–∞ –Ω–æ—Ä–º–∞–ª—å–Ω—ã—Ö —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–π –ø–æ –ö–æ–ª–º–æ–≥–æ—Ä–æ–≤—É. –í–∏–∑—É–∞–ª—å–Ω–æ –æ—Ü–µ–Ω–∏—Ç—å –¥–∞–Ω–Ω—ã–π —Ñ–∞–∫—Ç.

```python
from scipy.stats import chisquare, kstest, norm, chi2, stats, t
# –ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –Ω–∞–±–ª—é–¥–µ–Ω–∏–π
n = len(Y)
s_2 = np.sum((Y - Y_mean)**2) / (n - 1)

print("–ù–µ—Å–º–µ—â–µ–Ω–Ω–∞—è –æ—Ü–µ–Ω–∫–∞ –¥–∏—Å–ø–µ—Ä—Å–∏–∏:", s_2)
predicted_Y = beta_0 + beta_1 * X

# –í—ã—á–∏—Å–ª–∏–º –æ—à–∏–±–∫–∏
errors = Y - predicted_Y

h = 1  # –®–∞–≥ –¥–ª—è –≥–∏—Å—Ç–æ–≥—Ä–∞–º–º—ã
plt.figure(figsize=(10, 5))
plt.hist(errors, bins=np.arange(min(errors), max(errors) + h, h), edgecolor='black', alpha=0.7)
plt.title('–ì–∏—Å—Ç–æ–≥—Ä–∞–º–º–∞ –æ—à–∏–±–æ–∫')
plt.xlabel('–û—à–∏–±–∫–∏')
plt.ylabel('–ß–∞—Å—Ç–æ—Ç–∞')
plt.grid()
plt.show()

# –ü—Ä–æ–≤–µ—Ä–∏–º –≥–∏–ø–æ—Ç–µ–∑—É –Ω–æ—Ä–º–∞–ª—å–Ω–æ—Å—Ç–∏ –æ—à–∏–±–æ–∫ –Ω–∞ —É—Ä–æ–≤–Ω–µ –∑–Ω–∞—á–∏–º–æ—Å—Ç–∏ 0.1 —Å –ø–æ–º–æ—â—å—é –∫—Ä–∏—Ç–µ—Ä–∏—è —Ö–∏-–∫–≤–∞–¥—Ä–∞—Ç:
observed_frequencies, bin_edges = np.histogram(errors, bins=np.arange(min(errors), max(errors) + h, h)) #–ø–æ–¥—Å—á–µ—Ç —á–∞—Å—Ç–æ—Ç
expected_frequencies = n * np.diff(bin_edges) / np.sum(np.diff(bin_edges)) #–≤—ã—á–∏—Å–ª—è–µ–º –æ–∂–∏–¥–∞–µ–º—ã–µ —á–∞—Å—Ç–æ—Ç—ã
chi2_statistic = np.sum((observed_frequencies - expected_frequencies)**2 / expected_frequencies) #–≤—ã—á–∏—Å–ª–µ–Ω–∏–µ –∑–Ω–∞—á–µ–Ω–∏—è —Å—Ç–∞—Ç–∏—Å—Ç–∏–∫–∏ –∫—Ä–∏—Ç–µ—Ä–∏—è —Ö–∏-–∫–≤–∞–¥—Ä–∞—Ç
# –í—ã—á–∏—Å–ª—è–µ–º –∫—Ä–∏—Ç–∏—á–µ—Å–∫–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ —Ö–∏-–∫–≤–∞–¥—Ä–∞—Ç
chi2_critical_value = chi2.ppf(0.9, df=len(observed_frequencies) - 1)
print("–°—Ç–∞—Ç–∏—Å—Ç–∏–∫–∞ —Ö–∏-–∫–≤–∞–¥—Ä–∞—Ç",chi2_statistic)
print("–ö—Ä–∏—Ç–∏—á–µ—Å–∫–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ —Ö–∏-–∫–≤–∞–¥—Ä–∞—Ç",chi2_critical_value)

if chi2_statistic > chi2_critical_value:
    print('–û—à–∏–±–∫–∏ –Ω–µ –ø–æ–¥—á–∏–Ω—è—é—Ç—Å—è –Ω–æ—Ä–º–∞–ª—å–Ω–æ–º—É —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—é (–æ—Ç–∫–ª–æ–Ω–∏—Ç—å –Ω—É–ª–µ–≤—É—é –≥–∏–ø–æ—Ç–µ–∑—É)')
else:
    print('–û—à–∏–±–∫–∏ –º–æ–≥—É—Ç –∏–º–µ—Ç—å –Ω–æ—Ä–º–∞–ª—å–Ω–æ–µ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ (–Ω–µ –æ—Ç–≤–µ—Ä–≥–∞—Ç—å –Ω—É–ª–µ–≤—É—é –≥–∏–ø–æ—Ç–µ–∑—É).')

# –û—Ü–µ–Ω–∏–º —Ä–∞—Å—Å—Ç–æ—è–Ω–∏–µ –ø–æ–ª—É—á–µ–Ω–Ω–æ–π –æ—Ü–µ–Ω–∫–∏ –¥–æ –∫–ª–∞—Å—Å–∞ –Ω–æ—Ä–º–∞–ª—å–Ω—ã—Ö —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–π –ø–æ –ö–æ–ª–º–æ–≥–æ—Ä–æ–≤—É:
ks, crit_value = kstest(Y, 'norm')
print("–°—Ç–∞—Ç–∏—Å—Ç–∏–∫–∞ –ö–æ–ª–º–æ–≥–æ—Ä–æ–≤–∞:", ks)
print("p-–∑–Ω–∞—á–µ–Ω–∏–µ:", crit_value)
alpha = 0.02
if crit_value < alpha:
    print("–û—Ü–µ–Ω–∫–∞ –Ω–µ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤—É–µ—Ç –Ω–æ—Ä–º–∞–ª—å–Ω–æ–º—É —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—é.")
else:
    print("–û—Ü–µ–Ω–∫–∞ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤—É–µ—Ç –Ω–æ—Ä–º–∞–ª—å–Ω–æ–º—É —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—é.")

# –í–∏–∑—É–∞–ª—å–Ω–æ –æ—Ü–µ–Ω–∏–º –¥–∞–Ω–Ω–æ–µ —Ä–∞—Å—Ç–æ—è–Ω–∏–µ:
plt.hist(errors, bins=np.arange(min(errors), max(errors) + h, h), edgecolor='black', alpha=0.7)
plt.xlabel('Errors')
plt.ylabel('–ß–∞—Å—Ç–æ—Ç–∞')
plt.title('–ì–∏—Å—Ç–æ–≥—Ä–∞–º–º–∞ –æ—Ü–µ–Ω–∫–∏ –æ—à–∏–±–æ–∫ –≤–º–µ—Å—Ç–µ —Å —Ñ—É–Ω–∫—Ü–∏–µ–π –ø–ª–æ—Ç–Ω–æ—Å—Ç–∏ –Ω–æ—Ä–º–∞–ª—å–Ω–æ–≥–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è')
x = np.linspace(min(errors), max(errors), 1000)
plt.plot(x, norm.pdf(x, np.mean(errors), np.std(errors)), 'r', label='–ù–æ—Ä–º–∞–ª—å–Ω–æ–µ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ')
plt.legend()
plt.grid()
plt.show()
```

### 3.–í –ø—Ä–µ–¥–ø–æ–ª–æ–∂–µ–Ω–∏–∏ –Ω–æ—Ä–º–∞–ª—å–Ω–æ—Å—Ç–∏ –æ—à–∏–±–æ–∫ –ø–æ—Å—Ç—Ä–æ–∏—Ç—å –¥–æ–≤–µ—Ä–∏—Ç–µ–ª—å–Ω—ã–π –∏–Ω—Ç–µ—Ä–≤–∞–ª –¥–ª—è –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤ $\beta_0,\beta_1$ —É—Ä–æ–≤–Ω—è –¥–æ–≤–µ—Ä–∏—è $1-\alpha$.–ü–æ—Å—Ç—Ä–æ–∏—Ç—å –¥–æ–≤–µ—Ä–∏—Ç–µ–ª—å–Ω—ã–π —ç–ª–ª–∏–ø—Å —É—Ä–æ–≤–Ω—è –¥–æ–≤–µ—Ä–∏—è $1-\alpha$ –¥–ª—è ($\beta_0,\beta_1$) (–≤—ã—á–∏—Å–ª–∏—Ç—å –µ–≥–æ –ø–æ–ª—É–æ—Å–∏).

```python
# –í—ã—á–∏—Å–ª—è–µ–º –æ—Ü–µ–Ω–∫—É —Å—Ç–∞–Ω–¥–∞—Ä—Ç–Ω–æ–π –æ—à–∏–±–∫–∏ –¥–ª—è beta_0 –∏ beta_1
se_beta_0 = np.sqrt(s_2 * ((1/n) + (X_mean**2 / ((n - 1) * np.sum((X - X_mean)**2)))))
se_beta_1 = np.sqrt(s_2 / (np.sum((X - X_mean)**2)))
alpha = 0.1

# –í—ã—á–∏—Å–ª—è–µ–º –∫—Ä–∏—Ç–∏—á–µ—Å–∫–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ t-—Å—Ç–∞—Ç–∏—Å—Ç–∏–∫–∏
t_critical = -np.abs(t.ppf(alpha/2, n - 2))

# –í—ã—á–∏—Å–ª—è–µ–º –≥—Ä–∞–Ω–∏—Ü—ã –¥–æ–≤–µ—Ä–∏—Ç–µ–ª—å–Ω–æ–≥–æ –∏–Ω—Ç–µ—Ä–≤–∞–ª–∞
CI_beta_0 = [beta_0 - t_critical * se_beta_0, beta_0 + t_critical * se_beta_0]
CI_beta_1 = [beta_1 - t_critical * se_beta_1, beta_1 + t_critical * se_beta_1]

print("–î–æ–≤–µ—Ä–∏—Ç–µ–ª—å–Ω—ã–π –∏–Ω—Ç–µ—Ä–≤–∞–ª –¥–ª—è beta_0:", CI_beta_0)
print("–î–æ–≤–µ—Ä–∏—Ç–µ–ª—å–Ω—ã–π –∏–Ω—Ç–µ—Ä–≤–∞–ª –¥–ª—è beta_1:", CI_beta_1)
# –í—ã—á–∏—Å–ª—è–µ–º –∫–æ–≤–∞—Ä–∏–∞—Ü–∏–æ–Ω–Ω—É—é –º–∞—Ç—Ä–∏—Ü—É
beta_1 = np.repeat(beta_1, len(errors))
cov_matrix = np.cov(np.column_stack((errors, beta_1)))

# –í—ã—á–∏—Å–ª—è–µ–º —Å–æ–±—Å—Ç–≤–µ–Ω–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è –∏ —Å–æ–±—Å—Ç–≤–µ–Ω–Ω—ã–µ –≤–µ–∫—Ç–æ—Ä—ã
eigenvalues, eigenvectors = np.linalg.eig(cov_matrix)

# –í—ã—á–∏—Å–ª—è–µ–º –ø–æ–ª—É–æ—Å–∏ –¥–æ–≤–µ—Ä–∏—Ç–µ–ª—å–Ω–æ–≥–æ —ç–ª–ª–∏–ø—Å–∞
a = np.sqrt(eigenvalues[0] * chi2.ppf(1 - alpha/2, 2))
b = np.sqrt(eigenvalues[1] * chi2.ppf(1 - alpha/2, 2))

print("–ü–æ–ª—É–æ—Å—å a –¥–æ–≤–µ—Ä–∏—Ç–µ–ª—å–Ω–æ–≥–æ —ç–ª–ª–∏–ø—Å–∞:", a)
print("–ü–æ–ª—É–æ—Å—å b –¥–æ–≤–µ—Ä–∏—Ç–µ–ª—å–Ω–æ–≥–æ —ç–ª–ª–∏–ø—Å–∞:", b)
```

### 4.–°—Ñ–æ—Ä–º—É–ª–∏—Ä–æ–≤–∞—Ç—å –≥–∏–ø–æ—Ç–µ–∑—É –Ω–µ–∑–∞–≤–∏—Å–∏–º–æ–π –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π Y –æ—Ç –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π X. –ü—Ä–æ–≤–µ—Å—Ç–∏ –ø—Ä–æ–≤–µ—Ä–∫—É –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç–∏.

```python
from scipy.stats import f
p = 2  # —á–∏—Å–ª–æ –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤ –≤ –ª–∏–Ω–µ–π–Ω–æ–π —Ä–µ–≥—Ä–µ—Å—Å–∏–∏

# –í—ã–≤–æ–¥–∏–º —Å—É–º–º—É –∫–≤–∞–¥—Ä–∞—Ç–æ–≤ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–π, –æ–±—ä—è—Å–Ω–µ–Ω–Ω—É—é —Å –ø–æ–º–æ—â—å—é –º–æ–¥–µ–ª–∏
SSE = np.sum((Y - (beta_0 + beta_1 * X)) ** 2)

# –í—ã–≤–æ–¥–∏–º —Å—É–º–º—É –∫–≤–∞–¥—Ä–∞—Ç–æ–≤ –æ—Å—Ç–∞—Ç–∫–æ–≤, –æ—Å—Ç–∞–≤—à—É—é—Å—è –ø–æ—Å–ª–µ —Ä–∞–±–æ—Ç—ã –º–æ–¥–µ–ª–∏
SSR = np.sum((beta_0 + beta_1 * X - np.mean(Y)) ** 2)

# –í—ã–≤–æ–¥–∏–º –æ–±—â—É—é —Å—É–º–º—É –∫–≤–∞–¥—Ä–∞—Ç–æ–≤ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–π
SST = SSE + SSR

MSE = SSE / (n - p)  # —Å—Ä–µ–¥–Ω–µ–∫–≤–∞–¥—Ä–∞—Ç–∏—á–Ω–æ–µ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ –æ—Å—Ç–∞—Ç–∫–æ–≤
MSR = SSR / p  # —Å—Ä–µ–¥–Ω–µ–∫–≤–∞–¥—Ä–∞—Ç–∏—á–Ω–æ–µ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ –æ–±—ä—è—Å–Ω–µ–Ω–Ω—ã—Ö –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–π

F = MSR / MSE  # –≤—ã—á–∏—Å–ª—è–µ–º –∑–Ω–∞—á–µ–Ω–∏–µ F-–∫—Ä–∏—Ç–µ—Ä–∏—è

alpha = 0.1

F_critical = f.ppf(1 - alpha, p, n - p)  # –∫—Ä–∏—Ç–∏—á–µ—Å–∫–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ F-–∫—Ä–∏—Ç–µ—Ä–∏—è
print("F-–∫—Ä–∏—Ç–µ—Ä–∏–π",F)
print("–ö—Ä–∏—Ç–∏—á–µ—Å–∫–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ F = ",F_critical)
if F > F_critical:
    print("–û—Ç–≤–µ—Ä–≥–∞–µ–º –Ω—É–ª–µ–≤—É—é –≥–∏–ø–æ—Ç–µ–∑—É. –°—É—â–µ—Å—Ç–≤—É–µ—Ç –∑–Ω–∞—á–∏–º–∞—è –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç—å –º–µ–∂–¥—É Y –∏ X.")
else:
    print("–ù–µ –æ—Ç–≤–µ—Ä–≥–∞–µ–º –Ω—É–ª–µ–≤—É—é –≥–∏–ø–æ—Ç–µ–∑—É. –ù–µ—Ç –∑–Ω–∞—á–∏–º–æ–π –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç–∏ –º–µ–∂–¥—É Y –∏ X.")
```

### 5.–°—Ñ–æ—Ä–º—É–ª–∏—Ä–æ–≤–∞—Ç—å –º–æ–¥–µ–ª—å –≤–∫–ª—é—á–∞—é—â—É—é –¥–æ–ø–ª–Ω–∏—Ç–µ–ª—å–Ω—ã–π —á–ª–µ–Ω —Å $X^2$.–ü–æ—Å—Ç—Ä–æ–∏—Ç—å –ú–ù–ö –æ—Ü–µ–Ω–∫–∏ –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤ $\beta_0,\beta_1,\beta_2$ –≤ –¥–∞–Ω–Ω–æ–π –º–æ–¥–µ–ª–∏. –ò–∑–æ–±—Ä–∞–∑–∏—Ç—å –≥—Ä–∞—Ñ–∏—á–µ—Å–∫–∏ –ø–æ–ª—É—á–µ–Ω–Ω—É—é —Ä–µ–≥—Ä–µ—Å—Å–∏–æ–Ω–Ω—É—é –º–æ–¥–µ–ª—å.

–†–µ–≥—Ä–µ—Å—Å–∏–æ–Ω–Ω–∞—è –º–æ–¥–µ–ª—å –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π Y –ø–æ –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π X —Ñ–æ—Ä–º—É–ª–∏—Ä—É–µ—Ç—Å—è —Å–ª–µ–¥—É—é—â–∏–º –æ–±—Ä–∞–∑–æ–º:

$Y = \beta_0 + \beta_1X + \beta_2X^2 + \varepsilon,$

–≥–¥–µ $\beta_0,\beta_1,\beta_2$ - –ø–∞—Ä–∞–º–µ—Ç—Ä—ã, –∞ $\varepsilon$ - —Å–ª—É—á–∞–π–Ω–∞—è –æ—à–∏–±–∫–∞.


```python
Y = np.array([16.81, 3.45, 6.39, 1.67, 13.91, 1.11, 6.20, 17.13, 8.57, 9.23, 8.66, 6.01, 9.31, 7.64, 4.30, 5.28, 14.32,1.53, 16.19, 14.03, 21.10, 8.32, 3.65, 3.87, 19.21, 6.20, 19.03, 16.54, 11.21, 5.56, 9.17, 0.79, 9.66, 5.77,4.84, 12.58, 5.91, 1.93, 2.88, 1.02, 2.62, 7.64, 9.86, 6.59, 18.69, 0.81, 15.91, 2.68, 15.63, 14.85])
X = np.array([0, 4, 1, 3, 2, 4, 2, 3, 2, 2, 4, 4, 0, 1, 2, 3, 2, 4, 0, 3, 2, 2, 4, 0, 1, 3, 1, 0, 0, 0, 4, 0, 2, 3,0, 3, 4, 2, 1, 1, 1, 3, 0, 1, 4, 2, 1, 2, 1, 0])
X_2 = X ** 2
X_stacked = np.column_stack((np.ones_like(X), X, X_2))

# –ò—Å–ø–æ–ª—å–∑—É–µ–º –º–µ—Ç–æ–¥ –Ω–∞–∏–º–µ–Ω—å—à–∏—Ö –∫–≤–∞–¥—Ä–∞—Ç–æ–≤ –¥–ª—è –Ω–∞—Ö–æ–∂–¥–µ–Ω–∏—è –æ—Ü–µ–Ω–æ–∫ –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤
beta = np.linalg.lstsq(X_stacked, Y, rcond=None)[0]

beta_0 = beta[0]
beta_1 = beta[1]
beta_2 = beta[2]

print("–û—Ü–µ–Ω–∫–∏ –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤:")
print("Œ≤0 = ",beta_0)
print("Œ≤1 = ",beta_1)
print("Œ≤2 = ",beta_2)

x = np.linspace(min(X), max(X), 100)
y = beta_0 + beta_1 * x + beta_2 * x ** 2

plt.scatter(X, Y, label='–ò—Å—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ')
plt.plot(x, y, 'r', label='–†–µ–≥—Ä–µ—Å—Å–∏–æ–Ω–Ω–∞—è –º–æ–¥–µ–ª—å')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('–ú–æ–¥–µ–ª—å —Ä–µ–≥—Ä–µ—Å—Å–∏–∏ —Å $X^2$')
plt.legend()
plt.show()
```

### 6.–ü–æ—Å—Ç—Ä–æ–∏—Ç—å –Ω–µ—Å–º–µ—â–µ–Ω–Ω—É—é –æ—Ü–µ–Ω–∫—É –¥–∏—Å–ø–µ—Ä—Å–∏–∏. –ü—Ä–æ–≤–µ—Å—Ç–∏ –∏—Å—Å–ª–µ–¥–æ–≤–∞–Ω–∏–µ –Ω–æ—Ä–º–∞–ª—å–Ω–æ—Å—Ç–∏ –æ—à–∏–±–æ–∫ –∫–∞–∫ –≤ –ø—É–Ω–∫—Ç–µ 2.

```python
from scipy.stats import chisquare, kstest, norm, chi2
n = len(Y)
s_2 = np.sum((Y - Y_mean)**2) / (n - 1)

print("–ù–µ—Å–º–µ—â–µ–Ω–Ω–∞—è –æ—Ü–µ–Ω–∫–∞ –¥–∏—Å–ø–µ—Ä—Å–∏–∏:", s_2)
predicted_Y = beta_0 + beta_1 * X + beta_2 * X ** 2

errors = Y - predicted_Y

h = 1
plt.hist(errors, bins=np.arange(min(errors), max(errors) + h, h), edgecolor='black', alpha=0.7)
plt.xlabel('–û—à–∏–±–∫–∏')
plt.ylabel('–ß–∞—Å—Ç–æ—Ç–∞')
plt.title('–ì–∏—Å—Ç–æ–≥—Ä–∞–º–º–∞ –æ—Ü–µ–Ω–∫–∏ –æ—à–∏–±–æ–∫')
plt.show()

# –ü—Ä–æ–≤–µ—Ä–∫–∞ –≥–∏–ø–æ—Ç–µ–∑—ã –æ –Ω–æ—Ä–º–∞–ª—å–Ω–æ—Å—Ç–∏ –æ—à–∏–±–æ–∫ —Å –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏–µ–º –∫—Ä–∏—Ç–µ—Ä–∏—è —Ö–∏-–∫–≤–∞–¥—Ä–∞—Ç
observed_frequencies, bin_edges = np.histogram(errors, bins=np.arange(min(errors), max(errors) + h, h))
expected_frequencies = n * np.diff(bin_edges) / np.sum(np.diff(bin_edges))
chi2_statistic, _ = chisquare(observed_frequencies, expected_frequencies)

# –í—ã—á–∏—Å–ª–µ–Ω–∏–µ –∫—Ä–∏—Ç–∏—á–µ—Å–∫–æ–≥–æ –∑–Ω–∞—á–µ–Ω–∏—è —Ö–∏-–∫–≤–∞–¥—Ä–∞—Ç
chi2_critical_value = chi2.ppf(0.9, df=len(observed_frequencies) - 1)

print("–°—Ç–∞—Ç–∏—Å—Ç–∏–∫–∞ —Ö–∏-–∫–≤–∞–¥—Ä–∞—Ç:", chi2_statistic)
print("–ö—Ä–∏—Ç–∏—á–µ—Å–∫–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ —Ö–∏-–∫–≤–∞–¥—Ä–∞—Ç:", chi2_critical_value)

if chi2_statistic > chi2_critical_value:
    print('–û—à–∏–±–∫–∏ –Ω–µ –ø–æ–¥—á–∏–Ω—è—é—Ç—Å—è –Ω–æ—Ä–º–∞–ª—å–Ω–æ–º—É —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—é (–æ—Ç–≤–µ—Ä–≥–Ω—É—Ç—å –Ω—É–ª–µ–≤—É—é –≥–∏–ø–æ—Ç–µ–∑—É)')
else:
    print('–û—à–∏–±–∫–∏ –º–æ–≥—É—Ç –∏–º–µ—Ç—å –Ω–æ—Ä–º–∞–ª—å–Ω–æ–µ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ (–Ω–µ –æ—Ç–≤–µ—Ä–≥–∞—Ç—å –Ω—É–ª–µ–≤—É—é –≥–∏–ø–æ—Ç–µ–∑—É)')

# –û—Ü–µ–Ω–∫–∞ —Ä–∞—Å—Å—Ç–æ—è–Ω–∏—è –æ—Ü–µ–Ω–æ–∫ –¥–æ –Ω–æ—Ä–º–∞–ª—å–Ω–æ–≥–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –ø–æ –∫—Ä–∏—Ç–µ—Ä–∏—é –ö–æ–ª–º–æ–≥–æ—Ä–æ–≤–∞-–°–º–∏—Ä–Ω–æ–≤–∞
ks_statistic, ks_p_value = kstest(errors, 'norm')

print("–°—Ç–∞—Ç–∏—Å—Ç–∏–∫–∞ –ö–æ–ª–º–æ–≥–æ—Ä–æ–≤–∞-–°–º–∏—Ä–Ω–æ–≤–∞:", ks_statistic)
print("p-–∑–Ω–∞—á–µ–Ω–∏–µ:", ks_p_value)

if ks_p_value < alpha:
    print("–û—à–∏–±–∫–∏ –Ω–µ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤—É—é—Ç –Ω–æ—Ä–º–∞–ª—å–Ω–æ–º—É —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—é")
else:
    print("–û—à–∏–±–∫–∏ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤—É—é—Ç –Ω–æ—Ä–º–∞–ª—å–Ω–æ–º—É —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—é")

# –í–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è –æ—à–∏–±–æ–∫ –∏ –ø–ª–æ—Ç–Ω–æ—Å—Ç–∏ –Ω–æ—Ä–º–∞–ª—å–Ω–æ–≥–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è
plt.hist(errors, bins=np.arange(min(errors), max(errors) + h, h), edgecolor='black', density=True, alpha=0.7)
plt.plot(x, norm.pdf(x, np.mean(errors), np.std(errors)), 'r')
plt.xlabel('–û—à–∏–±–∫–∏')
plt.ylabel('–ü–ª–æ—Ç–Ω–æ—Å—Ç—å')
plt.title('–ì–∏—Å—Ç–æ–≥—Ä–∞–º–º–∞ –æ—à–∏–±–æ–∫ —Å –ø–ª–æ—Ç–Ω–æ—Å—Ç—å—é –Ω–æ—Ä–º–∞–ª—å–Ω–æ–≥–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è')
plt.show()
```

### 7.–í –ø—Ä–µ–¥–ø–æ–ª–æ–∂–µ–Ω–∏–∏ –Ω–æ—Ä–º–∞–ª—å–Ω–æ—Å—Ç–∏ –æ—à–∏–±–æ–∫ –ø–æ—Å—Ç—Ä–æ–∏—Ç—å –¥–æ–≤–µ—Ä–∏—Ç–µ–ª—å–Ω—ã–µ –∏–Ω—Ç–µ—Ä–≤–∞–ª—ã –¥–ª—è –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤ $\beta_0,\beta_1,\beta_2$ —É—Ä–æ–≤–Ω—è –¥–æ–≤–µ—Ä–∏—è $1-\alpha$. –ù–∞–ø–∏—Å–∞—Ç—å —É—Ä–∞–≤–Ω–µ–Ω–∏–µ –¥–æ–≤–µ—Ä–∏—Ç–µ–ª—å–Ω–æ–≥–æ —ç–ª–ª–∏–ø—Å–æ–∏–¥–∞ —É—Ä–æ–≤–Ω—è –¥–æ–≤–µ—Ä–∏—è $1-\alpha$.

```python
alpha = 0.1
n = len(X)

# –í—ã—á–∏—Å–ª—è–µ–º —Å—Ç–∞–Ω–¥–∞—Ä—Ç–Ω—É—é –æ—à–∏–±–∫—É –æ—Ü–µ–Ω–∫–∏
mse = np.sum(errors ** 2) / (n - 3)
cov_matrix = mse * np.linalg.inv(X_stacked.T @ X_stacked)
std_errors = np.sqrt(np.diag(cov_matrix))

# –í—ã—á–∏—Å–ª—è–µ–º t-—Å—Ç–∞—Ç–∏—Å—Ç–∏–∫—É –¥–ª—è —É—Ä–æ–≤–Ω—è –¥–æ–≤–µ—Ä–∏—è alpha/2
t_value = t.ppf(1 - alpha/2, n - 3)

# –í—ã—á–∏—Å–ª—è–µ–º –¥–æ–≤–µ—Ä–∏—Ç–µ–ª—å–Ω—ã–µ –∏–Ω—Ç–µ—Ä–≤–∞–ª—ã –¥–ª—è –∫–∞–∂–¥–æ–≥–æ –ø–∞—Ä–∞–º–µ—Ç—Ä–∞
confidence_intervals = np.column_stack((beta - t_value * std_errors, beta + t_value * std_errors))

print("–î–æ–≤–µ—Ä–∏—Ç–µ–ª—å–Ω—ã–µ –∏–Ω—Ç–µ—Ä–≤–∞–ª—ã:")
print(f"Œ≤0: [{confidence_intervals[0, 0]}, {confidence_intervals[0, 1]}]")
print(f"Œ≤1: [{confidence_intervals[1, 0]}, {confidence_intervals[1, 1]}]")
print(f"Œ≤2: [{confidence_intervals[2, 0]}, {confidence_intervals[2, 1]}]")
# –í—ã—á–∏—Å–ª—è–µ–º –∫—Ä–∏—Ç–∏—á–µ—Å–∫–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ –¥–ª—è F-—Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è
F_value = f.ppf(1 - alpha, 3, n - 3)

# –í—ã—á–∏—Å–ª—è–µ–º –ø–æ–ª—É–æ—Å–∏ —ç–ª–ª–∏–ø—Å–æ–∏–¥–∞
semi_axes = np.sqrt(F_value) * np.sqrt(np.diag(cov_matrix))

print("–£—Ä–∞–≤–Ω–µ–Ω–∏–µ –¥–æ–≤–µ—Ä–∏—Ç–µ–ª—å–Ω–æ–≥–æ —ç–ª–ª–∏–ø—Å–æ–∏–¥–∞:")
print(f"(({beta[1]} - x1) / {semi_axes[1]})**2 + (({beta[2]} - x2) / {semi_axes[2]})**2 <= 1")
```

### 8.–°—Ñ–æ—Ä–º—É–ª–∏—Ä–æ–≤–∞—Ç—å –≥–∏–ø–æ—Ç–µ–∑—É –ª–∏–Ω–µ–π–Ω–æ–π —Ä–µ–≥—Ä–µ—Å—Å–∏–æ–Ω–Ω–æ–π –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç–∏ –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π Y –æ—Ç –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π X –∏ –ø—Ä–æ–≤–µ—Ä–∏—Ç—å –µ—ë –∑–Ω–∞—á–∏–º–æ—Å—Ç—å –Ω–∞ —É—Ä–æ–≤–Ω–µ $\alpha$.


–ì–∏–ø–æ—Ç–µ–∑–∞ –ª–∏–Ω–µ–π–Ω–æ–π —Ä–µ–≥—Ä–µ—Å—Å–∏–æ–Ω–Ω–æ–π –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç–∏ –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π Y –æ—Ç –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π X –º–æ–∂–µ—Ç –±—ã—Ç—å —Å—Ñ–æ—Ä–º—É–ª–∏—Ä–æ–≤–∞–Ω–∞ —Å–ª–µ–¥—É—é—â–∏–º –æ–±—Ä–∞–∑–æ–º:

H0: $\beta_1 = \beta_2 = 0$ (–Ω–µ—Ç –ª–∏–Ω–µ–π–Ω–æ–π –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç–∏ –º–µ–∂–¥—É Y –∏ X)

H1: –•–æ—Ç—è –±—ã –æ–¥–∏–Ω –∏–∑ –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç–æ–≤ $\beta_1, \beta_2$ –Ω–µ —Ä–∞–≤–µ–Ω –Ω—É–ª—é (—Å—É—â–µ—Å—Ç–≤—É–µ—Ç –ª–∏–Ω–µ–π–Ω–∞—è –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç—å)

```python
# –í—ã–≤–æ–¥–∏–º —Å—É–º–º—É –∫–≤–∞–¥—Ä–∞—Ç–æ–≤ –æ—Å—Ç–∞—Ç–∫–æ–≤, –æ—Å—Ç–∞–≤—à—É—é—Å—è –ø–æ—Å–ª–µ —Ä–∞–±–æ—Ç—ã –º–æ–¥–µ–ª–∏
SSR = np.sum((beta_0 + beta_1 * X + beta_2 * X ** 2 - np.mean(Y)) ** 2)
# –í—ã–≤–æ–¥–∏–º —Å—É–º–º—É –∫–≤–∞–¥—Ä–∞—Ç–æ–≤ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–π, –æ–±—ä—è—Å–Ω–µ–Ω–Ω—É—é —Å –ø–æ–º–æ—â—å—é –º–æ–¥–µ–ª–∏
SSE = np.sum((Y - (beta_0 + beta_1 * X + beta_2 * X ** 2)) ** 2)
p = 3
# –í—ã–≤–æ–¥–∏–º –æ–±—â—É—é —Å—É–º–º—É –∫–≤–∞–¥—Ä–∞—Ç–æ–≤ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–π
SST = SSE + SSR
MSE = SSE / (n - p)  # —Å—Ä–µ–¥–Ω–µ–∫–≤–∞–¥—Ä–∞—Ç–∏—á–Ω–æ–µ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ –æ—Å—Ç–∞—Ç–∫–æ–≤
MSR = SSR / p  # —Å—Ä–µ–¥–Ω–µ–∫–≤–∞–¥—Ä–∞—Ç–∏—á–Ω–æ–µ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ –æ–±—ä—è—Å–Ω–µ–Ω–Ω—ã—Ö –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–π
F = MSR / MSE  # –≤—ã—á–∏—Å–ª—è–µ–º –∑–Ω–∞—á–µ–Ω–∏–µ F-–∫—Ä–∏—Ç–µ—Ä–∏—è

alpha = 0.1

F_critical = f.ppf(1 - alpha, p, n - p)  # –∫—Ä–∏—Ç–∏—á–µ—Å–∫–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ F-–∫—Ä–∏—Ç–µ—Ä–∏—è

print("F =",F)
print("–ö—Ä–∏—Ç–∏—á–µ—Å–∫–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ =", F_critical)
# –ü—Ä–æ–≤–µ—Ä–∫–∞ –∑–Ω–∞—á–∏–º–æ—Å—Ç–∏ –≥–∏–ø–æ—Ç–µ–∑—ã
if F > F_critical:
    print("–ì–∏–ø–æ—Ç–µ–∑–∞ –ª–∏–Ω–µ–π–Ω–æ–π —Ä–µ–≥—Ä–µ—Å—Å–∏–æ–Ω–Ω–æ–π –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç–∏ –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π Y –æ—Ç –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π X –∑–Ω–∞—á–∏–º–∞ –Ω–∞ —É—Ä–æ–≤–Ω–µ 0.1.")
else:
    print("–ì–∏–ø–æ—Ç–µ–∑–∞ –ª–∏–Ω–µ–π–Ω–æ–π —Ä–µ–≥—Ä–µ—Å—Å–∏–æ–Ω–Ω–æ–π –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç–∏ –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π Y –æ—Ç –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π X –Ω–µ –∑–Ω–∞—á–∏–º–∞ –Ω–∞ —É—Ä–æ–≤–Ω–µ 0.1.")
```

## –í—ã–≤–æ–¥—ã
–í –ø—Ä–æ—Ü–µ—Å—Å–µ –≤—ã–ø–æ–ª–Ω–µ–Ω–∏—è –ø—Ä–∞–∫—Ç–∏—á–µ—Å–∫–æ–π —Ä–∞–±–æ—Ç—ã –±—ã–ª–∏ —Ä–∞—Å—Å–º–æ—Ç—Ä–µ–Ω—ã —Å–ª–µ–¥—É—é—â–∏–µ –∞—Å–ø–µ–∫—Ç—ã:

–†–µ–≥—Ä–µ—Å—Å–∏–æ–Ω–Ω—ã–π –∞–Ω–∞–ª–∏–∑, –∫–æ—Ç–æ—Ä—ã–π –∏–≥—Ä–∞–µ—Ç –∫–ª—é—á–µ–≤—É—é —Ä–æ–ª—å –∫–∞–∫ –≤ —Å—Ç–∞—Ç–∏—Å—Ç–∏–∫–µ, —Ç–∞–∫ –∏ –≤ –º–∞—à–∏–Ω–Ω–æ–º –æ–±—É—á–µ–Ω–∏–∏. –†–∞–±–æ—Ç–∞ –≤–∫–ª—é—á–∞–ª–∞ —Å–æ–∑–¥–∞–Ω–∏–µ –º–æ–¥–µ–ª–∏, –æ—Ü–µ–Ω–∫—É –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤ –∏ –∞–Ω–∞–ª–∏–∑ –∏—Ö –∑–Ω–∞—á–∏–º–æ—Å—Ç–∏, –ø—Ä–æ–≤–µ—Ä–∫—É –≥–∏–ø–æ—Ç–µ–∑—ã –æ –Ω–µ–∑–∞–≤–∏—Å–∏–º–æ—Å—Ç–∏ –∏ –ª–∏–Ω–µ–π–Ω–æ–π —Ä–µ–≥—Ä–µ—Å—Å–∏–æ–Ω–Ω–æ–π –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç–∏, –∞ —Ç–∞–∫–∂–µ –æ—Ü–µ–Ω–∫—É –Ω–æ—Ä–º–∞–ª—å–Ω–æ—Å—Ç–∏ –æ—à–∏–±–æ–∫ –∏ –ø–æ—Å—Ç—Ä–æ–µ–Ω–∏–µ –¥–æ–≤–µ—Ä–∏—Ç–µ–ª—å–Ω—ã—Ö –∏–Ω—Ç–µ—Ä–≤–∞–ª–æ–≤ –∏ —ç–ª–ª–∏–ø—Å–æ–≤.
